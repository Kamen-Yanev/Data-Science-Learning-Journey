The exercise goes as follows: 
Write a code that counts the number of unique vowels for a word the user provides and returns the count. The uppercase variant of the same vowel does not count as unique.

```python
word = input('Enter word:') #input returns a string so it works as is
vowels = ['a', 'e', 'i', 'o', 'u']
vowel_count = [] #will use as empty list to receive unique values and count them
for letter in word.lower():
    if letter in vowels and letter not in vowel_count:
        vowel_count.append(letter)
vowel_suffix = 'vowel' if len(vowel_count) == 1 else 'vowels'
print(f'{word} has {len(vowel_count)} unique {vowel_suffix}!')
```

Exercise 2: Write a program that classifies user input as one of the following:
- empty input
- a number
- a lowercase word
- an uppercase word
- a mixed-case word
- a word containing non-letter characters

```python
word = input('Enter any word:')
if word == '':
    print('Your input is empty.')
elif word.replace('.', '', 1).isdigit(): #This removes the decimal point and treats the number as an int
    print('That\'s a number!')
elif word.startswith('-') and word[1:].replace('.', '', 1).isdigit(): #This handles negative numbers
    print('That\'s a negative number!')
elif word.isalpha(): #This checks if the word consists of alphabetic characters only
    if word == word.upper():
        print('That\'s an uppercase word!')
    elif word == word.lower():
        print('That\'s a lowercase word!')
    else:
        print('That\'s a mixed-case word!')
else:
    print('Your word contains special characters!')
```

Exercise 3: Simple Password Validator

A password is considered valid if:
1. It is at least 8 characters long
2. It contains at least one letter
3. It contains at least one number
4. It contains at least one special character
5. It does not contain spaces

```python
special_characters =['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '-', '+', '=', '?', ':']
password = input('Enter password:')
if len(password)<8:
    print('Your password must be at least 8 characters long.')
elif not any(character.isalpha() for character in password):
    print('Your password must contain at least one letter.')
elif not any(character.isdigit() for character in password):
    print('Your password must contain at least one number.')
elif not any(character in special_characters for character in password):
    print('''Your password must contain at least one special character from the following list:
    !,@,#,$,%,^,&,*,(, ), -, +, =, ?''')
elif ' ' in password:
    print('Your password must not contain spaces.')
else:
    print('Password is valid')
```
